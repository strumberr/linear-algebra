```python
import numpy as np

def LL_product(L1, L2):
    
    n = len(L1)
    C = np.zeros((n, n))

    for i in range(n):
        for j in range(i + 1):
            C[i][j] = sum(L1[i][k] * L2[k][j] for k in range(j + 1))
    
    print(C)
    return C
```
