```python

import numpy as np

def permutation_matrix(perm):
    # YOUR CODE HERE
    N = len(perm)
    I = np.eye(N)
    

    P = np.zeros((N, N))
    
    for i in range(N):
        P[i] = I[perm[i]]
    
    return P
```
