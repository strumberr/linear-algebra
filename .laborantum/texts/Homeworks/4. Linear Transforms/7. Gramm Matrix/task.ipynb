```python
import numpy as np

def contra_to_co(B, x):
   
    G = np.dot(B.T, B)
    
    x_dot_B = np.dot(B.T, x)
    x_covariant = np.linalg.solve(G, x_dot_B)
    
    return x_covariant


```
